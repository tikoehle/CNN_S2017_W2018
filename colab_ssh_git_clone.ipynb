# Colab Github and Drive usage

Run a notebook in colab. Any imports are loaded from the git cloned repo in colab. Ssh is used for git cloning the repo to avoid entering a pw.

* **Create ssh keys for colab to ssh git clone from github**
```
ssh-keygen -f $HOME/.ssh/colab
cd .ssh
cat > config <<EOF
Host github.com
    IdentityFile /root/.ssh/colab
EOF
tar cvfz ssh_colab.tar.gz colab colab.pub config
```
* **Add the colab.pub key to github**
* **Run cell 1:**
  - Copy the ssh_colab.tar.gz to colab
  - git clone the project; essentially the import directories for running the notebooks
  - Load the datasets into the cloned colab repo; or do it from the notebook
